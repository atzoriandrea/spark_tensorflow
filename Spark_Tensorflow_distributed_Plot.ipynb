```python
import matplotlib.pyplot as plt
import numpy as np
```

**PLOT TEMPO TRAINING E PREDICT IN BASE AL NUMERO DI NODI DEL CLUSTER**



SVM Attribute Eval 100

```python
Num_nodi_cluster = [1,2,3,4,5,6,7,8]
Tempo_training = [72,204,216,258,282,302,363,329]
Tempo_predict = [27,28,29,29,32,32,30,34]

Tempo = Tempo_training + Tempo_predict
print(Tempo)

plt.figure(figsize=(25,10))
plt.plot(Num_nodi_cluster, Tempo_training, marker='o', linestyle='dashed', linewidth=2, markersize=6, color="red", label='Training')
plt.plot(Num_nodi_cluster, Tempo_predict, marker='*', linestyle='dashed', linewidth=2, markersize=6, color="blue", label='Predict')
plt.grid(color='b', linestyle='-', linewidth=0.1)
plt.xticks(Num_nodi_cluster, rotation=0)
plt.yticks(Tempo, rotation=0)
plt.ylabel("Tempo (sec)", fontsize=13)
plt.xlabel("Numero Nodi Cluster", fontsize=13)
plt.legend(fancybox=True, borderpad=1, loc="upper right", prop={'size': 15})
```
